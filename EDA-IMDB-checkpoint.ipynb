### Classification Of movies (Excellent,Good, Average)

```python
def classification(Rating) :
    if Rating>=8.0:
       return'Excellent'
    elif Rating>=7.0:
       return 'Good'
    else :
       return 'Average' 
```

```python
df['class']=df['Rating'].apply(classification)
```

### Eda on Each Genres 

```python
df['Genre']
```

```python
list1=[]
for item in df['Genre'] :
    list1.append (item.split(','))
     
```

```python
one_d=[]
for item in list1 :
    for item1 in item :
        one_d.append(item1)
```

```python
uni=[]
for item in one_d :
    if item not in uni:
        uni.append(item)
    
```

```python
uni
```

```python
from collections import Counter
```

```python
Counter(one_d)
```

```python
hit_genres=pd.DataFrame(one_d,columns=['Genres'])
```

```python
top_10_genres=hit_genres['Genres'].value_counts().sort_values(ascending=False).head(10)
```

```python
top_10_genres
```

```python
import matplotlib.pyplot as plt

# Colors for each segment
colors = ['skyblue', 'lightcoral', 'lightgreen', 'lightsalmon']

# Plotting the donut chart
plt.pie(top_10_genres, labels=top_10_genres.index, colors=colors, startangle=90, wedgeprops=dict(width=0.4))

# Draw a circle in the center to create the donut shape
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')

# Display the plot
plt.title('Hit Genres')
plt.show()
```
